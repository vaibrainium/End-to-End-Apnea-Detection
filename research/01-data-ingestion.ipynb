```python
from dataclasses import dataclass
from pathlib import Path

@dataclass(frozen=True)
class DataIngestionConfig:
    root_dir: Path
    raw_data_dir: Path
    ingested_dir: Path
```

```python
from apnea_detection import logger
from apena_detection.constants import CONFIG_FILE_PATH, PARAMS_FILE_PATH
from apena_detection.utils import read_yaml, create_directories

class ConfigurationManager:
    def __init__(self, config_file_path: CONFIG_FILE_PATH, params_file_path: PARAMS_FILE_PATH):
        self.config = read_yaml(config_file_path)
        self.params = read_yaml(params_file_path)
        create_directories([self.config.artifacts_root])

    def get_data_ingestion_config(self) -> DataIngestionConfig:
        config = self.config.data_ingestion
        create_directories([config.ingested_dir])
        data_ingestion_config = DataIngestionConfig(
            root_dir=Path(config.root_dir),
            raw_data_dir=Path(config.raw_data_dir),
            ingested_dir=Path(config.ingested_dir)
        )
        logger.info(f"Data Ingestion Config: {data_ingestion_config}")
        return data_ingestion_config
```
